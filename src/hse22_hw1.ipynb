```python
def get_lenghts(filename):
    lenghts = []
    with open(filename) as file:
        for line in file:
            if line[0] == '>':
                lenght = int(line.split('_')[1][3:])
                lenghts.append(lenght)
    return lenghts


def find_n50(lenghts):
    pref_sum = 0
    total_sum = sum(lenghts)
    for lenght in reversed(sorted(lenghts)):
        pref_sum += lenght
        if pref_sum >= total_sum / 2:
            return lenght


def get_longest_scaffold(filename):
    longest_scaffold = ''
    current_scaffold = ''
    with open(filename) as file:
        for line in file:
            if line[0] == '>':
                if len(current_scaffold) > len(longest_scaffold):
                    longest_scaffold = current_scaffold
                current_scaffold = ''
            else:
                current_scaffold += line.strip()
    return longest_scaffold


def get_gaps_amount_and_total_lenght(scaffold):
    # Кол-во гэпов = (кол-во 'N') - (кол-во 'NN' с учетом перекрытий)
    countN = scaffold.count('N')
    countNN = scaffold.split('N').count('')
    return countN - countNN, countN
```

Анализ контигов

```python
contig_lenghts = get_lenghts('contigs_contig.fa')
print('Общее кол-во контигов:', len(contig_lenghts))
print('Общая длина всех контигов:', sum(contig_lenghts))
print('Длина самого длинного контига:', max(contig_lenghts))
print('N50:', find_n50(contig_lenghts))
```

Анализ скаффолдов

```python
scaffold_lenghts = get_lenghts('contigs_scaffold.fa')
print('Общее кол-во скаффолдов:', len(scaffold_lenghts))
print('Общая длина всех скаффолдов:', sum(scaffold_lenghts))
print('Длина самого длинного скаффолда:', max(scaffold_lenghts))
print('N50:', find_n50(scaffold_lenghts))
```

Анализируем гэпы в самом длинном скаффолде

```python
longest_scaffold = get_longest_scaffold('contigs_scaffold.fa')
gap_amount, gap_total_lenght = get_gaps_amount_and_total_lenght(longest_scaffold)

print('Число гэпов:', gap_amount)
print('Общая длина гэпов:', gap_total_lenght)
```

Для скаффолдов с уменьшенным числом гэпов

```python
longest_scaffold_gapclosed = get_longest_scaffold('contigs_gapClosed.fa')
gap_amount, gap_total_lenght = get_gaps_amount_and_total_lenght(longest_scaffold_gapclosed)

print('Число гэпов (gapClosed):', gap_amount)
print('Общая длина гэпов (gapClosed):', gap_total_lenght)
```

Сохраняем самый длинный скаффолд

```python
with open('longest.fa', 'w') as file:
    file.write(longest_scaffold_gapclosed)
```
